```python
#backward() does the differentiation and forgrad = true should be set or else the varialbe wont be calculated
```

```python
x = torch.tensor(3.)
w = torch.tensor(4.,requires_grad=True)
b = torch.tensor(5.,requires_grad = True)
x, w, b
```

```python
y = w*x + b
```

### ðŸ”¹ requires_grad = True (PyTorch)

When we set `requires_grad=True` for a tensor, PyTorch starts tracking all the operations performed on it.  
So, if we perform a complex operation involving multiple tensors (each having `requires_grad=True`),  
PyTorch can automatically calculate the **partial derivatives** of the final output with respect to **each of those variables** during backpropagation.  
In short, it allows PyTorch to apply **automatic differentiation** and compute gradients for all variables involved.


```python
y.backward()
```

```python
print(x.grad)#since we did not set requires grad = true it wont be keeping in the memory the values of x to calculate gradients
print(b.grad) #, now its only 3 variables but when they are millions this would be crazy optimization
print(w.grad)
```
